## 🧠 Reinforcement Learning as a Story Over Time

Think of Reinforcement Learning as a process that unfolds step by step over time, like a story or a game being played.

At every moment, an **agent** (the learner or decision-maker) interacts with an **environment**.

The interaction follows a simple loop:

- The agent is in a **state** (it represents the current situation or context).
- The agent takes an **action** (it makes a decision or performs a move).
- The environment responds with a **reward** (a numerical signal indicating how good or bad the action was).
- The environment then transitions to a **new state** (the situation changes, and the cycle repeats).

This can be summarized as a continuous feedback loop:

> **State → Action → Reward → New State → (repeat)**

This repeating cycle is the foundation of all Reinforcement Learning problems.

## 🟦 State (S, St, s)

A **state** represents the current situation of the agent in the environment.

It answers the question:
> “Where am I right now?”

### Intuition:
Think of a state as a **snapshot of the world at a particular moment**.

For example:
- In a maze: your position (cell coordinates)
- In a game: current screen configuration
- In driving: speed, position, traffic conditions

### Notation:
- **S** → random variable representing a state
- **St** → state at time step *t*
- **s** → a specific state value

👉 A state contains all the information needed to make a decision.

## 🎮 Action (A, At, a)

An **action** is what the agent chooses to do in a given state.

It answers the question:
> “What should I do next?”

### Intuition:
Think of actions as **controls or moves** the agent can take.

Examples:
- Move left / right / up / down in a grid world
- Accelerate / brake in driving
- Pick an option in a game

### Notation:
- **A** → random variable representing an action
- **At** → action taken at time step *t*
- **a** → a specific action

👉 Actions change the state of the environment.

## 🏆 Reward (R, Rt)

A **reward** is a numerical signal from the environment that tells the agent how good or bad its action was.

It answers the question:
> “How well did I do?”

### Intuition:
Think of reward as **instant feedback**:
- Positive reward → good action
- Negative reward → bad action
- Zero reward → neutral action

Examples:
- +1 for winning a game
- -1 for losing
- 0 for each step in a maze

### Notation:
- **R** → random variable for reward
- **Rt** → reward received at time step *t*

👉 Reward is the only direct learning signal in Reinforcement Learning.

## ⏳ Time (t, T)

Reinforcement Learning happens over **discrete time steps**.

At each time step, the agent:
1. observes a state
2. takes an action
3. receives a reward
4. moves to the next state

### Notation:
- **t** → current time step (0, 1, 2, 3, ...)
- **T** → final time step (end of an episode)

### Episode:
An **episode** is a complete sequence of interaction:
> start → actions → rewards → termination

Example:
- One full game of chess
- One full maze run

👉 RL is about learning across many time steps, not just one decision.

## 🧭 Policy (π)

A **policy** is the agent’s decision-making strategy.

It defines:
> “What action should I take in a given state?”

### Intuition:
Think of policy as the **brain or behavior rule** of the agent.

It maps states → actions.

---

### 🟦 Deterministic Policy

A deterministic policy always chooses the same action for a given state:

$$
\pi(s) = a
$$
Meaning:
> “In state s, always take action a.”

✔ No randomness  
✔ Same input → same output  

---

### 🟩 Stochastic Policy

A stochastic policy gives probabilities over actions:

$$
\pi(a|s)
$$

Meaning:
> “In state s, take action a with some probability.”

Example:
- Left: 70%
- Right: 30%

✔ Includes randomness  
✔ Useful for exploration  

---

👉 Policy is what the agent is trying to improve during learning.

## 🌍 Dynamics of the Environment

The **environment dynamics** describe how the world responds to the agent’s actions.

It answers:
> “If I take an action in a state, what happens next?”

---

### Notation:
$$
p(s', r \mid s, a)
$$

This means:
> Probability of moving to next state `s'` and receiving reward `r`, given current state `s` and action `a`.

---

### Intuition:
Think of it as the **rules of the world**.

Example:
- In a game:
  - pressing a button may lead to different outcomes
- In a maze:
  - moving up may sometimes fail due to walls

---

### Key Idea:
The environment is often:
- stochastic (random)
- unknown to the agent
- something the agent must learn about indirectly

---

👉 Dynamics define how actions influence the future.

## 📈 Value Function vπ(s)

The **state-value function** tells us how good a state is.

It answers:
> “If I am in state *s* and follow policy π, how much reward can I expect in the long run?”

---

### 🧠 Intuition

Think of it like:
> “If I start from this situation and keep behaving according to my strategy, how good will things turn out?”

So it is not about immediate reward, but **future total reward**.

---

### 📌 Notation
- **vπ(s)** = value of state *s* under policy π

---

### 🎮 Example

Imagine a game:
- State A: near treasure → high value
- State B: near danger → low value

Even if both give same immediate reward, their future outcomes differ.

---

👉 Value function = long-term goodness of a state.

## 🏆 Optimal Value Functions (v*, q*)

These describe the **best possible performance** an agent can achieve.

---

### 🌟 Optimal State Value

$$
v^*(s)
$$

> “What is the maximum expected reward I can get starting from state s?”

This assumes:
- perfect behavior
- best possible policy

---

### 🌟 Optimal Action Value

$$
q^*(s, a)
$$

> “What is the best possible long-term reward if I take action a in state s?”

---

### 🧠 Intuition

Think of it like:
- v*(s) → best possible life outcome from this situation
- q*(s,a) → best possible outcome if I take this specific decision

---

👉 If we know q*(s,a), we can always act optimally by choosing the best action.

## ⚙️ Approximate Value Functions (v̂, q̂)

In real problems, we cannot store exact values for every state.

So we use approximation.

---

### 📌 Notation

- **v̂(s, w)** → approximate value of state s
- **q̂(s, a, w)** → approximate value of state-action pair

where:
- **w** = weight vector (parameters of a model)

---

### 🧠 Intuition

Think of this like:
> “Instead of memorizing values for every situation, we use a function (like a neural network) to estimate them.”

---

### 🎮 Example

A game has millions of states:
- impossible to store all v(s)
- so we approximate using features + weights

---

👉 This is how modern deep reinforcement learning works.

## 🧠 Features and Linear Value Approximation

To estimate values, we often use features.

---

### 📌 Notation

- **x(s)** → feature vector of state s
- **w** → weight vector
- **w · x(s)** → dot product

---

### 🧠 Intuition

We represent a state using numbers:
- distance to goal
- health level
- position coordinates

Then we combine them:

$$
\hat{v}(s, w) = w^T x(s)
$$

Meaning:
> “Value = weighted sum of features”

---

### 🎮 Example

If:
- feature1 = closeness to goal
- feature2 = danger level

Then:
- higher closeness → higher value
- higher danger → lower value

---

👉 Features are how the agent “understands” the world numerically.

## 📊 Return (Gt) — Total Future Reward

The **return** is the total amount of reward the agent expects to receive from a time step onward.

---

### 📌 Definition

$$
G_t = R_{t+1} + R_{t+2} + R_{t+3} + \cdots
$$

But usually we use **discounting**:

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots
$$

---

### 🧠 Intuition

Think of return as:
> “How much total reward will I collect in the future from this point?”

But:
- immediate reward matters more
- far future reward matters less

---

### 🎮 Example

If you get rewards:
- step 1: +1
- step 2: +1
- step 3: +10

Return = total of all future rewards (with discounting)

---

👉 Return = long-term score of a trajectory.

## 🎯 Discount Factor (γ)

The **discount factor γ** controls how much the agent cares about the future.

---

### 📌 Range

$$
0 \leq \gamma \leq 1
$$

---

### 🧠 Intuition

Think of γ as:
> “How far into the future should I care?”

---

### Cases:

- **γ = 0**
  - only immediate reward matters
  - very short-sighted agent

- **γ close to 1 (e.g., 0.99)**
  - future rewards are almost as important as immediate ones
  - long-term planning

---

### 🎮 Example

- Eating candy now vs getting health benefits later:
  - low γ → eat candy
  - high γ → choose health

---

👉 γ shapes the agent’s “patience”.

## ⚡ Temporal Difference Error (δt)

The **TD error** measures how wrong the agent’s prediction was.

---

### 📌 Idea

$$
\delta_t = \text{target} - \text{prediction}
$$

---

### In RL form:

$$
\delta_t = R_{t+1} + \gamma V(S_{t+1}) - V(S_t)
$$

---

### 🧠 Intuition

Think of it like:

> “I expected this state to be worth X, but after seeing what happened, I realize it should be X + correction.”

---

### 🎮 Example

- You thought a move would be good (value = 5)
- After playing, you see it leads to worse outcomes (target = 2)

So:
- δt is negative → you overestimated

---

👉 TD error = learning signal that fixes predictions.

## 🧠 Eligibility Trace (et)

Eligibility traces help the agent decide:

> “Which past states/actions should get credit or blame?”

---

### 📌 Intuition

Think of it like a **memory signal**:
- recent events = strong influence
- older events = fading influence

---

### 🧠 Analogy

Imagine you dropped a ball:
- which earlier decisions caused the outcome?
- eligibility trace assigns “responsibility levels”

---

### 📉 Decay idea

Recent steps matter more than older ones.

---

👉 Eligibility traces help connect actions to delayed rewards.

## 🎲 ε-greedy Policy (Exploration vs Exploitation)

The ε-greedy policy balances:

- **Exploitation** → choose best known action
- **Exploration** → try random actions

---

### 📌 Definition

- With probability **1 − ε** → choose best action
- With probability **ε** → choose random action

---

### 🧠 Intuition

Think:
> “Sometimes I trust my experience, sometimes I try new things.”

---

### 🎮 Example

- ε = 0.1:
  - 90% best known move
  - 10% random move

---

### Why needed?

Without exploration:
- agent may get stuck in bad behavior forever

---

👉 ε-greedy ensures learning never stops exploring.
